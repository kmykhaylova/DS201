## Problem 1
The spam data has four columns which are all features variables. We do not need a dependent y variable for the clustering problem. Use the KMeans algorithm to find 2 clusters of the data using all four columns. Next, make predictions using the KMeans and find predictions of the labels. Display the cluster centers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics as me
from sklearn.cluster import KMeans as km
from sklearn import preprocessing as pre
import scipy.stats as st
```

```python
spam = pd.read_csv('spam.csv')
spam.columns
```

```python
x = spam[['x1', 'x2', 'x3', 'x4']]
```

```python
kmeans = km(n_clusters = 2)
kmeans.fit(x)
ykmeans = kmeans.predict(x)
ykmeans
```

```python
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'black', s = 100, alpha = 0.7)
```

### Problem 2
Following the model in problem 1, make a scatter plot of 2 clusters ( in different colors) with their centers. You can pick a pair of columns on the x and y-axis that clearly shows the separate clusters.

```python
x = np.array(x)
plt.scatter(x[:, 0], x[:, 1], c = ykmeans, s = 10, cmap = 'viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'black', s = 100, alpha = 0.7)
```
