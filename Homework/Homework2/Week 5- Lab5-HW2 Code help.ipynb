```python
newvg['weighted_sales']= 0.5 * newvg.NA_Sales+ 0.2 * newvg.EU_Sales +0.2 * newvg.JP_Sales +0.1 * newvg.Other_Sales
```

```python
newvg.head()
```

```python
newvg['scaled_sales']= newvg['weighted_sales']/max(newvg['weighted_sales'])*5
```
