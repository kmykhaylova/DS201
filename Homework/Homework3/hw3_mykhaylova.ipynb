### Problem 1
Denote the data set with all the columns except "Column 0" and "rate" as $x$. Then denote the "rate" variable with $y$. Train a Linear Regression model with $y$ as a dependent variable and $x$ as a set of independent variables. Make sure to add constant-coefficient in $x$ before running the model. Report the summary, $R^2$ and all the variables with the p-value of  $p>0.10$.

```python
import pandas as pd
import numpy as np
import matplotlib as mpt
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
data = pd.read_csv('Highway1-2.csv')
data.columns
```

```python
y = data['rate']
```

```python
x = data[['len', 'ADT', 'trks', 'sigs1', 'slim', 'shld', 'lane', 'acpt', 'itg', 'lwid']]
x = sm.add_constant(x)
```

```python
data.head()
```

```python
model = sm.OLS(y,x)
result = model.fit()
result.summary()
```

R^2 is 0.743. Variables with P > 0.9 are ADT and lane.
